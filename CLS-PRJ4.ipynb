## Rectangle Class:Define a class Rectangle with:
 ○ Two attributes: length and width.
 ○ A method area() to calculate the area.
 ○ A method perimeter() to calculate the perimeter.

```python
# Define the Rectangle class
class Rectangle:
    # Constructor to initialize the length and width of the rectangle
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Method to calculate the area of the rectangle
    def area(self):
        return self.length * self.width

    # Method to calculate the perimeter of the rectangle
    def perimeter(self):
        return 2 * (self.length + self.width)

# Input: Get the length and width from the user
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

# Create a Rectangle object
rect = Rectangle(length, width)

# Output the area and perimeter of the rectangle
print(f"Area of the rectangle: {rect.area()}")
print(f"Perimeter of the rectangle: {rect.perimeter()}")
```

## Student Class: Create a class Student with:
 ○ Attributes: name, roll_number, and marks.
 ○ A method display_details() to print the student's details.
 ○ A method is_passed() that returns True if marks ≥ 40, else
 False.

```python
# Define the Student class
class Student:
    # Constructor to initialize name, roll_number, and marks
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    # Method to display the student's details
    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Marks: {self.marks}")

    # Method to check if the student has passed
    def is_passed(self):
        return self.marks >= 40  # Return True if marks are 40 or more, else False

# Input: Get the student's details from the user
name = input("Enter the student's name: ")
roll_number = input("Enter the student's roll number: ")
marks = float(input("Enter the student's marks: "))

# Create a Student object
student = Student(name, roll_number, marks)

# Display student's details
student.display_details()

# Check if the student has passed and display the result
if student.is_passed():
    print("Status: Passed")
else:
    print("Status: Failed")
```

##  Define a class Circle with:
 ● Attribute: radius.
 ● Methods: area() to calculate the area, circumference() to
 calculate the circumference.

```python
import math  # Import the math module for mathematical calculations

# Define the Circle class
class Circle:
    # Constructor to initialize the radius of the circle
    def __init__(self, radius):
        self.radius = radius

    # Method to calculate the area of the circle
    def area(self):
        return math.pi * self.radius ** 2  # Area formula: π * r^2

    # Method to calculate the circumference of the circle
    def circumference(self):
        return 2 * math.pi * self.radius  # Circumference formula: 2 * π * r

# Input: Get the radius of the circle from the user
radius = float(input("Enter the radius of the circle: "))

# Create a Circle object
circle = Circle(radius)

# Output the area and circumference of the circle
print(f"Area of the circle: {circle.area()}")
print(f"Circumference of the circle: {circle.circumference()}")
```

##  Create a class BankAccount with:
 ● Attributes: account_number, account_holder, balance.
 ● Methods:
 ○ deposit(amount) to add money to the account.
○ withdraw(amount) to subtract money (if sufficient
 balance is available).
 ○ display_balance() to print the current balance.


```python
# Define the BankAccount class
class BankAccount:
    # Constructor to initialize account number, account holder, and balance
    def __init__(self, account_number, account_holder, balance=0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    # Method to deposit money into the account
    def deposit(self, amount):
        if amount > 0:  # Ensure that the deposit amount is positive
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money from the account
    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:  # Ensure there is enough balance
                self.balance -= amount
                print(f"Withdrew {amount}. New balance: {self.balance}")
            else:
                print("Insufficient balance to withdraw the amount.")
        else:
            print("Withdrawal amount must be positive.")

    # Method to display the current balance
    def display_balance(self):
        print(f"Current balance: {self.balance}")

# Input: Get account details from the user
account_number = input("Enter account number: ")
account_holder = input("Enter account holder name: ")

# Create a BankAccount object with an initial balance of 0
account = BankAccount(account_number, account_holder)

# Display account balance (initial balance is 0)
account.display_balance()

# Input: Deposit money into the account
deposit_amount = float(input("Enter the deposit amount: "))
account.deposit(deposit_amount)

# Input: Withdraw money from the account
withdraw_amount = float(input("Enter the withdrawal amount: "))
account.withdraw(withdraw_amount)

# Display updated account balance
account.display_balance()
```

## A class Book with:
 ● Attributes: title, author, price.
 ● Methods:
 ○ apply_discount(discount) to reduce the price by a
 given percentage.
 ○ display_details() to print the book details.

```python
# Define the Book class
class Book:
    # Constructor to initialize title, author, and price
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    # Method to apply a discount to reduce the price by a given percentage
    def apply_discount(self, discount):
        if 0 < discount < 100:  # Ensure the discount is a valid percentage (between 0 and 100)
            discount_amount = (self.price * discount) / 100
            self.price -= discount_amount
            print(f"Discount of {discount}% applied. New price: {self.price}")
        else:
            print("Invalid discount percentage. It should be between 0 and 100.")

    # Method to display the book details
    def display_details(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Price: {self.price}")

# Input: Get book details from the user
title = input("Enter the book title: ")
author = input("Enter the book author: ")
price = float(input("Enter the book price: "))

# Create a Book object
book = Book(title, author, price)

# Display book details
book.display_details()

# Input: Apply a discount to the book price
discount = float(input("Enter the discount percentage: "))
book.apply_discount(discount)

# Display updated book details
book.display_details()
```

##  Create a class Calculator with:
 ○ A method add(a, b) to add two numbers.
 ○ A method subtract(a, b) to subtract two numbers.
 ○ A method multiply(a, b) to multiply two numbers.
 ○ A method divide(a, b) to divide two numbers (handle division
 by zero gracefully).

```python
# Define the Calculator class
class Calculator:
    # Method to add two numbers
    def add(self, a, b):
        return a + b

    # Method to subtract two numbers
    def subtract(self, a, b):
        return a - b

    # Method to multiply two numbers
    def multiply(self, a, b):
        return a * b

    # Method to divide two numbers (handling division by zero)
    def divide(self, a, b):
        if b == 0:
            return "Error: Division by zero is not allowed."
        else:
            return a / b

# Input: Get two numbers from the user
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

# Create a Calculator object
calc = Calculator()

# Perform calculations
print(f"Addition: {calc.add(a, b)}")
print(f"Subtraction: {calc.subtract(a, b)}")
print(f"Multiplication: {calc.multiply(a, b)}")
print(f"Division: {calc.divide(a, b)}")
```

##  Write a class Person with:
 ○ Attributes: name, age, gender.
 ○ Method introduce() to print a message like: "Hello, my name
 is [name], I am [age] years old, and I am a [gender]."
 ○ Create a subclass Employee that adds:
 ■ Attribute: job_title.
 ■ Override introduce() to include the job title.


```python
# Define the Person class
class Person:
    # Constructor to initialize name, age, and gender
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    # Method to introduce the person
    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, and I am a {self.gender}.")

# Define the Employee subclass that extends Person
class Employee(Person):
    # Constructor to initialize name, age, gender, and job_title
    def __init__(self, name, age, gender, job_title):
        super().__init__(name, age, gender)  # Call the constructor of the parent class
        self.job_title = job_title

    # Override the introduce method to include job title
    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, I am a {self.gender}, and I work as a {self.job_title}.")

# Input: Get person details from the user
name = input("Enter your name: ")
age = int(input("Enter your age: "))
gender = input("Enter your gender: ")

# Create a Person object
person = Person(name, age, gender)

# Introduce the person
person.introduce()

# Input: Get employee details from the user
job_title = input("Enter your job title: ")

# Create an Employee object
employee = Employee(name, age, gender, job_title)

# Introduce the employee
employee.introduce()
```

##  File Creation and Writing: Write a Python script to:
 ○ Create a file named data.txt.
 ○ Write the numbers from 1 to 10 (each on a new line) to the
 file.
 ○ Read the file and print its contents.

```python
# Create and open the file in write mode ('w')
with open('data.txt', 'w') as file:
    # Write the numbers from 1 to 10, each on a new line
    for i in range(1, 11):
        file.write(f"{i}\n")

# Read the file and print its contents
with open('data.txt', 'r') as file:
    contents = file.read()
    print(contents)
```

##  Word Count in File: Write a program that reads a file named sample.txt, counts the number of words in it, and prints the
 result.

```python
# Open the file 'sample.txt' in read mode
with open('sample.txt', 'r') as file:
    # Read the content of the file
    content = file.read()
    
    # Split the content into words and count them
    words = content.split()
    word_count = len(words)
    
    # Print the word count
    print(f"The number of words in the file is: {word_count}")
```
